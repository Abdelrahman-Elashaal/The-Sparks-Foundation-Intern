```python
# Importing libraries will be used

import pandas as pd
import numpy
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Importing the data set from built in datasets in sklearn library

from sklearn.datasets import load_iris
iris=load_iris()
```

```python
# Put the data into a dataframe

iris_df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
# Read the first 10 rows of the data

iris_df.head(10)
```

```python
# Get information about the data

iris_df.info()
```

```python
# Get the number of clusters using elbow method
# The KMeans algorithm requires the input data to be in the form of a numeric array or matrix

x=iris_df.values
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k,n_init=10)
    km.fit(x)
    sse.append(km.inertia_)
```

```python
# Graph the elbow method

plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

#### It seems the number of clusters is = 3

```python
# Fitting the model

km = KMeans(n_clusters=3,n_init=10)
y_kmeans = km.fit_predict(x)
y_kmeans
```

```python
# Visualising the clusters - On the first two columns (Sepal Columns)

plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 75, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 75, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 75, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters

plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:,1], 
            s = 50, c = 'yellow', label = 'Centroids')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend() ;
```

```python
# Visualising the clusters - On the Last two columns (Petal Columns)

plt.scatter(x[y_kmeans == 0, 2], x[y_kmeans == 0, 3], 
            s = 75, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 2], x[y_kmeans == 1, 3], 
            s = 75, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 2], x[y_kmeans == 2, 3],
            s = 75, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters

plt.scatter(km.cluster_centers_[:, 2], km.cluster_centers_[:,3], 
            s = 50, c = 'yellow', label = 'Centroids')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend() ;
```
